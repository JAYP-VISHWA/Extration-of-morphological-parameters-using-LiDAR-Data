```
# File must be using 
# uploaded=files.upload() 
# To upload files
# filename='filename.csv'
def csv_to_list_xyz(filename):
  str1=uploaded[filename]
  s=str1.decode('utf-8')
  t=s.split('\n')
  #print(t[1])
  #print(t)
  ord_x=[]
  ord_y=[]
  ord_z=[]
  for i in t:
    t1=i.split(' ')
    if len(t1)==3:
      ord_x.append(t1[0])
      ord_y.append(t1[1])
      ord_z.append(t1[2])
    #ord_x=np.array([float(j) for  j in t1[1]])
    #ord_x=np.array([float(j) for  j in t1[2]])
    #ord_x=np.array([float(k) for j,k,l in i.split(' ')])
    #ord_x=np.array([float(l) for j,k,l in i.split(' ')])
  #ord_x=np.array([float(i) for i in t[5::4]])
  #ord_y=np.array([float(i) for i in t[6::4]])
  #ord_z=np.array([float(i) for i in t[7::4]])
  ord_x=np.array([float(j) for j in ord_x])
  ord_y=np.array([float(j) for j in ord_y])
  ord_z=np.array([float(j) for j in ord_z])
  X=[[x,y,z] for x,y,z in zip(ord_x,ord_y,ord_z)]
  return ord_x,ord_y,ord_z,X
```

```
X_ground_x,X_ground_y,X_ground_z,X_ground=csv_to_list_xyz(filename='text.txt')
```

```
print(X_ground)
```

```
good_val=[[i,j,k] for i,j,k in X_ground if (k>=30 and k<=36)]
X_good_x=[i for i,j,k in good_val]
X_good_y=[j for i,j,k in good_val]
X_good_z=[k for i,j,k in good_val]
```

```
from mpl_toolkits import mplot3d
%matplotlib inline
import matplotlib.pyplot as plt
```

```
fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection='3d')
ax.scatter3D(X_ground_x, X_ground_y, X_ground_z, color = "green",s=1)
plt.title("3D scatter plot-DBSCAN Tuning")
ax.set_xlabel('EPS', fontweight ='bold') 
ax.set_ylabel('Min_samples', fontweight ='bold') 
ax.set_zlabel('Number_of_Clusters', fontweight ='bold')

 
# show plot
plt.show()
```

```
fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection='3d')
ax.scatter3D(X_good_x, X_good_y, X_good_z, color = "green",s=1)
plt.title("3D scatter plot-DBSCAN Tuning: Good values")
ax.set_xlabel('EPS', fontweight ='bold') 
ax.set_ylabel('Min_samples', fontweight ='bold') 
ax.set_zlabel('Number_of_Clusters', fontweight ='bold')

 
# show plot
plt.show()
```
